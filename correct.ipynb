```python
start=0
listYear=[]
check=df['Date'][0]
for k in range(1,len(df['Date'])):
    if df['Date'][k]>check:
        check=df['Date'][k]
    elif df['Date'][k]<check:
        end=k-1
        listYear.append([start,end])
        start=k
        check=df['Date'][k]
listYear.append([start,k])
listYear
```

```python
year1=2023
for k in range(1,len(listYear)):
    for m in range(listYear[k][1]-listYear[k][0]+1):
        df.iloc[listYear[k][0]+m,0]=df.iloc[listYear[k][0]+m,0].replace(year=year1)
    year1+=1
```
